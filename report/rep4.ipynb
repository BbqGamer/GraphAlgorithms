```python
#EVALUATION OF HAMILTON
df = pd.read_csv("../data/csv/hamiltonNIGHT.csv")
df['time'] = df['time']/1000
p = pd.pivot_table(df, values="time", index="n", columns="k", aggfunc="mean")
p.plot(
    title="Time to find hamiltonian cycle",
    ylabel="Time (ms)",
    xlabel="Number of vertices",
    figsize=(8, 5),
    grid=True
)
```

```python
#EVALUATION OF HAMILTON
df = pd.read_csv("../data/csv/hamilton.csv")
df['time'] = df['time']/1000
df = df[df['k'] > 0.4]
p = pd.pivot_table(df, values="time", index="n", columns="k", aggfunc="mean")
p.plot(
    title="Time to find hamilton cycle",
    ylabel="Time (ms)",
    xlabel="Number of vertices",
    figsize=(8, 5),
    grid=True
)
```
